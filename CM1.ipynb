```python
## IMPORT PACKAGES
import numpy as np
import csv
import keras  
import tensorflow as tf 
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from sklearn.metrics import accuracy_score as ascore

# IMPORT TRAINING DATA 2008 (MATLAB-REFINED)
with open('trainingDataFinal.csv','r') as bv_data: 
    data_iter = csv.reader(bv_data, delimiter = ',', quotechar = '"') 
    data = [data for data in data_iter] 
    data = data[0:len(data)]                            
    data_array = np.asarray(data, dtype = np.float32)
    
#ADDITIONAL DATA PREPROCESING
#we transform the y labels from scalar values to one-hot encodings.
N_total = 64667;
d = 382; #using MATLAB preprocessing
N_train = round(N_total/2)

dNew = 7;
np.random.shuffle(data_array)
X_train = data_array[1:N_train, 0:d];
Y_train = data_array[1:N_train, d]
X_val = data_array[N_train+1:N_total, 0:d]
Y_val = data_array[N_train+1:N_total, d]
Y_train.astype(int)

resultsVec = [[0,0]]*9

#PARAMETERS TO TEST
wclassVec = [1.,5.,20.]

# NEURAL NETWORK MODEL FOR 3 LAYERS (DEPTH = 3)
widthVec = [[125,50,25], [67, 67, 66], [25, 50, 125]]
for w  in range(0,3):
    for c in range(0,3):
        #model creation
        model = Sequential()
        model.add(Dense(widthVec[w][0], input_dim=d))
        model.add(Dropout(0.7)) #does not count
        model.add(Dense(widthVec[w][1], input_dim=d))
        model.add(Dense(widthVec[w][2], input_dim=d))
        model.add(Dense(1)) #does not count
        model.add(Activation('sigmoid'))

        #Priorities class = 2 over other classes.
        class_weight = {0 : 1.,
        1: wclassVec[c]}
        ## Printing a summary of the layers and weights in your model
        model.summary()

        #Specifies the loss and optimizer
        model.compile(loss='binary_crossentropy',optimizer='rmsprop', metrics=['accuracy'])

        #Fit the data - specify batch size and number of epochs
        fit = model.fit(X_train, Y_train, batch_size=128, nb_epoch=20, class_weight = class_weight,
        verbose=1)

        ## Printing/save the accuracy of our model, according to the loss function specified in model.compile above
        trainscore = model.evaluate(X_train, Y_train, verbose=0)
        print('Train score:', trainscore[0])
        print('Train accuracy:', trainscore[1])
        valscore = model.evaluate(X_val, Y_val, verbose=0)
        print('Validation score:', valscore[0])
        print('Validation accuracy:', valscore[1])
   
        resultsVec[int(3*(w) + c)][0] = trainscore
        resultsVec[int(3*(w) + c)][1] = valscore
```
